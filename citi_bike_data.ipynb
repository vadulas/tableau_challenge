```python
import pandas as pd

years = ['2018','2019', '2020']
#years = ['2017']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

cumulative_data = pd.DataFrame()
#Loop through the years to append data
for year in years:
    
    for month in months:
        resource_file  = f"Resources/JC-{year}{month}-citibike-tripdata.csv" 
        print(resource_file)
        data_df = pd.read_csv(resource_file)
        
        #Insert a year-month column
        data_df.insert(1, 'year_month',  f'{year}{month}')
        data_df['year_month'] = pd.to_datetime(f'{year}-{month}')
        
        #Ensure that all the column names are all lower case
        data_df= data_df.rename(columns=str.lower)
        
        columns  = data_df.columns
        print(columns)

        #Clean up the column names for consistency
        if 'trip duration' in columns:
            data_df = data_df.rename(columns={'trip duration': 'tripduration'})
        if 'start time' in columns:
            data_df = data_df.rename(columns={'start time': 'starttime'})
        if 'stop time' in columns:
            data_df = data_df.rename(columns={'stop time': 'stoptime'})
        if 'bikeid' in columns:
            data_df = data_df.rename(columns={'bikeid': 'bike id'})
        if 'user type' in columns:
            data_df = data_df.rename(columns={'user type': 'usertype'})
        
        #Drop the rows where there is any missing data
        data_df = data_df.dropna()
        cumulative_data = cumulative_data.append(data_df)
        

cumulative_data.to_csv("Resources/all_data2.csv", index=False)
cumulative_data.columns 
```
